```python
df=covid19_confirmed.groupby('Country/Region').sum().reset_index()
df=df.sort_values(confirmed_latest,ascending=False).head(30).drop(['Lat', 'Long'],axis=1).set_index("Country/Region").T
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
df.plot(ax=ax,figsize=(20,20))
ax.set_title(f'5 COVID-19 Time series Confirmed as of {confirmed_latest}',fontsize= 30)
plt.xlabel('Date',fontsize= 20)
plt.ylabel('Number',fontsize= 20)
plt.rc('legend', fontsize=16)
plt.tick_params(labelsize=16)
fig.autofmt_xdate() # mke space for and rotate the x-axis tick labels
plt.legend(loc=2)
plt.show()
```

```python
# Total Confirmed by date
from matplotlib.ticker import ScalarFormatter

covid19_confirmed.loc["Total"] = covid19_confirmed.sum()
df=covid19_confirmed.iloc[-1,4:]
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
df.plot(ax=ax,figsize=(20,20))
ax.set_title(f'COVID-19 Time series Global confirmed as of {confirmed_latest}',fontsize= 30)
plt.xlabel('Date',fontsize= 20)
plt.ylabel('Number',fontsize= 20)
plt.tick_params(labelsize=16)
ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.ticklabel_format(style="plain",  axis="y")
fig.autofmt_xdate() # mke space for and rotate the x-axis tick labels
plt.show()
```
